```python
import matplotlib.pyplot as plt
months =[1,2,3,4]
sales=[1000,1500,1200,1800]
plt.xlabel("months")
plt.ylabel("sales")
plt.title("Monthly sales data report")

plt.plot(months,sales,color='red',linestyle='--',linewidth=2,marker='o',label='2025 sales data')

plt.show()
```

```python
import matplotlib.pyplot as plt
months =[1,2,3,4]
sales=[1000,1500,1200,1800]
plt.xlabel("months")
plt.ylabel("sales")
plt.title("Monthly sales data report")

plt.plot(months,sales,color='red',linestyle='--',linewidth=2,marker='o',label='2025 sales data')
plt.legend(loc='upper left',fontsize=12)
plt.grid(color='gray',linestyle=':',linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

#Bar chart

```python
product =['A','B','C','D']
sales=[1000,1500,800,1200]

plt.bar(product,sales,color='yellow',label='Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()
plt.show()
```
