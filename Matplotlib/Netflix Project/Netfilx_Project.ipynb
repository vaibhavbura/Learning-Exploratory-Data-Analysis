```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the data
df=pd.read_csv('/content/netflix_titles.csv')
df
```

```python
df.isnull()
```

```python
df.info()
```

```python
#clean data
df =df.dropna(subset=['type','release_year','rating','country','duration']) #removing nan values
df
```

```python
type_counts=df['type'].value_counts()
type_counts
```

```python
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=['skyblue','orange'])
plt.title('Number of Movies VS TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('movies_vs_tvshows.png')
plt.show()
```

```python
rating_count = df['rating'].value_counts()
rating_count
filtered_rating_count = rating_count[rating_count >= 250]
filtered_rating_count
```

```python
plt.figure(figsize=(8,6))
plt.pie(rating_count,labels=rating_count.index,autopct="%1.1f%%",startangle=90)
plt.title('Percentage of Content Rating')
plt.tight_layout()
plt.savefig('percentage of content rating')
plt.show()
```

```python
movie_df=df[df['type']=='Movie'].copy()
movie_df
```

```python
movie_df['duration_int'] = movie_df['duration'].str.replace('min','').astype(int)
movie_df['duration_int']
```

```python
plt.figure(figsize=(8,6))
plt.hist(movie_df['duration_int'],bins=30,color='purple',edgecolor='black')
plt.title("Distribution of Movie Durtion")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number pf Movies")
plt.tight_layout()
plt.savefig("movie_duration_fig.png")
plt.show()
```

```python
#release year vs number of shows using scatter plot
shows_df=df[df['type']=='TV Show'].copy()
shows_df
```

```python
#release_counts=df['release_year'].value_counts().sort_index()
release_counts = shows_df['release_year'].value_counts().sort_index()

release_counts
```

```python
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index,release_counts.values,color='red')
plt.title("Release year vs number of Shows")
plt.xlabel('Release Year')
plt.ylabel('Number of Shows')
plt.tight_layout()
plt.savefig("release year scatter.png")
plt.show()
```

```python
country_counts=shows_df['country'].value_counts().head(10)
country_counts
```

```python
plt.figure(figsize=(12,8))
plt.bar(country_counts.index,country_counts.values,color='teal')
plt.xlabel('Top 10 countries by Number of Shows')
plt.ylabel('Number of shows')
plt.tight_layout()
plt.savefig("top10 countries by shows")
plt.show()
```

```python
#subplot
content_by_year=df.groupby(['release_year','type']).size().unstack().fillna(0)
content_by_year
```

```python
fig, ax =plt.subplots(1,2,figsize=(12,5))

#first subplot:movies

ax[0].plot(content_by_year.index,content_by_year['Movie'],color='red')
ax[0].set_title("Movies Released Per Year")
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Number of movies')

#second subplot: TV Show
ax[1].plot(content_by_year.index,content_by_year['TV Show'],color='blue')
ax[1].set_title("TV Shows Released Per Year")
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Number of shows')
plt.tight_layout()
plt.savefig("movie tv shows release subplot")
plt.show()
```
