#Bar chart

```python
product =['A','B','C','D']
sales=[1000,1500,800,1200]

plt.bar(product,sales,color='yellow',label='Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()
plt.show()
```

#Pie Chart

```python
import matplotlib.pyplot as plt

regions=['North','South','East','West']

revenue=[3000,2000,1500,1000]

plt.pie(revenue,labels=regions,autopct='%1.1f',colors=['gold','skyblue','lightgreen','coral'])
plt.title('Revenue Contribution by region')
plt.show()

```

#Histogram

```python

scores=[10,54,65,14,68,55,54,68,90,35,21,15,55,65,47,55,46,12,68,45,56]
plt.hist(scores,bins=5,color='green',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of students')
plt.title('Score Distribution of Students')
plt.show()
```

# Scatter Plot

```python
hours_study=[1,2,3,4,5,6,7,8]

exam_score=[50,55,60,65,55,70,80,90]

plt.scatter(hours_study,exam_score,color='orange',marker='d',label='Student Data')
plt.xlabel('Hours studied')
plt.ylabel('Exam Score')
plt.title('Relationship between study and exam score')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.scatter([1,2,3],[50,55,60],color='blue',label='Class A')#g1
plt.scatter([1,2,3],[65,100,75],color='orange',label='Class B')#g2

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title("Comparison of two classes")
plt.legend()
plt.grid(True)
plt.show()
```

#SubPlot

```python
x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1) #1 row ,#2column,1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)#1 row ,#2column,2st subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
plt.savefig('save.png',dpi=300,bbox_inches='tight')
```

```python
x=[1,2,3,4]
y=[10,20,15,25]

#fig,ax=plt.subplots[nrows,ncols,figsize=(width,height)]
fig ,ax =plt.subplots(1,2,figsize=(10,5))

x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title('Line Plot')


ax[1].bar(x,y)
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of Line and Bar Charts')
plt.tight_layout()


plt.savefig('save.png',dpi=300,bbox_inches='tight')
plt.show()
```
